```python
from rng.spot import sample_spot_bars
import duckdb
```

```python
exchanges = ["drbt", "okex"]
bases = ["btc", "eth"]
quotes = ["usd", "usdc"]


spot_bars = sample_spot_bars(
    exchanges=exchanges,
    base_assets=bases,
    quote_assets=quotes,
    spread=0.001,
    seed=42,
)
```

```python
spot_bars.show()
```

```python
duckdb.register("spot_bars", spot_bars)
duckdb.query("SELECT * FROM spot_bars")
```
