```python
import matplotlib.pyplot as plt
```

```python
from rng import gbm

# 1-D: single GBM path
path = gbm(s0=100.0, mu=0.05, sigma=0.2, seed=42)
print(f"1-D path: shape={path.shape}, S[-1]={path[-1]:.4f}")
```

```python
plt.plot(path)
```

```python
import numpy as np
from rng import gbm

# 2-D: two correlated GBM paths
cov = np.array([[0.04, 0.02], [0.02, 0.09]])
paths = gbm(s0=[100.0, 50.0], mu=[0.05, 0.08], sigma=cov, seed=42)
print(f"2-D paths: shape={paths.shape}, S[-1]={paths[-1].round(4)}")
```

```python
plt.plot(paths)
```
